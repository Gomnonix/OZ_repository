<span style="font-size:80%">[문제6]<br><br></span>
```Python에서 메모리 관리는 어떻게 이루어지는지 설명해주세요```

<span style="color:red">답변</span>

Python은 자동으로 메모리를 관리하는 고급 프로그래밍언어이다.
Python의 메모리 관리는 참조 계수 referencecounting 기반으로 이루어진다.
이는 객체가 생성될 때마다 그 객체를 참조하는 변수의 개수를 추적하는 방식이다.
객체의 변수의 개수가 0이되면, 해당 객체의 메모리는 자동으로 해제된다.
