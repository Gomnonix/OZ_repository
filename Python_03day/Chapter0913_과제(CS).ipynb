1. 리스트 변경이 가능하다. 리스트의 요소는 추가, 삭제, 변경이 가능하다.

[문제3]

Python의 list는 연결리스트(linked list) 인가요?

리스트는 연결리스트가 아니다.
연결리스트란 다른 추상 자료형을 구현할 때 기반이 되는 기초 선형 자료 구조이다.
각 노드가 데이터와 포인터를 가지고 한 줄로 연결되어 잇는 방식으로 데이터를 저장한다.

[문제4]

딕셔너리 자료형에 대해 설명해주세요

딕셔너리의 자료형은 키와 값의 요소를 가지는 자료형이다.
중괄호 안에 키와 값을 요소로 생성한다. 요소를 생성한 키 변경이 어려우며 값의 변경을 가능하다.
